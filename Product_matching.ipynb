# Data Cleaning

```python
# Dropping columns from Flipkart & Amazon dataset, Which are unnecessary for the data analysis.


flipkart = flipkart_data.drop(columns=['crawl_timestamp', 'product_url','product_category_tree', 'pid','image',
                                 'is_FK_Advantage_product','overall_rating','description','product_specifications'])


amazon = amzon_data.drop(columns=['crawl_timestamp', 'product_url','product_category_tree', 'pid','image',
                                 'is_FK_Advantage_product','overall_rating','description','product_specifications'])
```

```python
# Checking actual rating values in Flipkart dataset
flipkart["product_rating"].unique()

flipkart["product_rating"].value_counts()
```

```python
# Checking actual rating values in Amazon dataset
amazon["product_rating"].unique()

amazon["product_rating"].value_counts()
```

```python
#checking for percentage of "No rating available" in product rating

flipkart_No_rating_available_percentage=(18151/20000)*100
amazon_No_rating_available_percentage=(18151/20000)*100


flipkart_No_rating_available_percentage
amazon_No_rating_available_percentage
```

```python
# Flipkart & Amazon both dataset has 90.75 % of null values with "no rating available",
# Dropping the 'product_rating' column in both the datasets due to the maximum number of null values.

flipkart.drop(columns=['product_rating'], inplace=True)

amazon.drop(columns=['product_rating'], inplace=True)
```

```python
# Checking for discounted price that are more than their retail price value

print((flipkart['retail_price'] < flipkart['discounted_price']).sum())
print((amazon['retail_price'] < amazon['discounted_price']).sum())
```

```python
# Their are 4325 value of discounted price which are greater than the retail_price, 
#dropping them

flipkart_wrong_val = flipkart[flipkart['retail_price'] < flipkart['discounted_price']].index
amazon_wrong_val = amazon[amazon['retail_price'] < amazon['discounted_price']].index

# Droping values that are greater than the retail price
amazon = amazon.drop(amazon_wrong_val)
flipkart = flipkart.drop(flipkart_wrong_val)
```

```python
# At sales flipkart and amazon dont give more than 80% Off on "retail_price",
#checking the discount values

print("Amazon: ",(amazon['discounted_price']/amazon['retail_price'] > 0.80).sum())
print("Flipkart: ",(flipkart['discounted_price']/flipkart['retail_price'] > 0.80).sum())
```

```python
# Removing data that has more than 80% discounted value.

over_discounted_Flipkart = flipkart[flipkart['discounted_price']/flipkart['retail_price'] > 0.80].index
over_discounted_Amazon = amazon[amazon['discounted_price']/amazon['retail_price'] > 0.80].index

```

```python
# Flipkart 
flipkart = flipkart.drop(over_discounted_Flipkart)

# Amazon 
amazon = amazon.drop(over_discounted_Amazon)
```

```python
# Checking null values for brands in both the dataset.

print("Amazon Brands: ",amazon['brand'].isna().sum())
print("Flipkart Brands: ",flipkart['brand'].isna().sum())
```

```python
# dropping the  null values from both dataset.

flipkart.dropna(inplace=True)
amazon.dropna(inplace=True)
```

```python
# renaming with apropriate columns for data analysis.

flipkart = flipkart.rename(columns={'product_name':'Flipkart_product_name','retail_price':'Flipkart_retail_price',
                                    'discounted_price':'Flipkart_discounted_price','brand':'Brand_in_Flipkart'})
amazon = amazon.rename(columns={'product_name':'Amazon_product_name','retail_price':'Amazon_retail_price',
                                    'discounted_price':'Amazon_discounted_price','brand':'Brand_in_Amazon'})
```

#Merging Flipkart and amazon dataset

```python
# Merging both the datasets by outer join.

product_data = pd.merge(flipkart,amazon, how='outer')
```

```python
# Creating new column for difference of discount price in Flipkart & Amazon 

product_data['Discount_price_diff'] = product_data['Flipkart_discounted_price'] - product_data['Amazon_discounted_price']
```
